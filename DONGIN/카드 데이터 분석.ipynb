```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import koreanize_matplotlib
import seaborn as sns
```

```python
data = pd.read_csv('BankChurners.csv')
data.columns
```

```python
data = data.drop(data.columns[[21,22]], axis=1)
```

```python
data.columns
```

```python
data.isnull().sum()
```

```python
data.head()
```

```python
industry_delivery_totals = data.groupby('Customer_Age')['CLIENTNUM'].sum().sort_values(ascending=False)

# Plotting the bar graph
plt.figure(figsize=(12, 4))
# industry_delivery_totals.plot.bar(x=industry_delivery_totals.index, y=industry_delivery_totals.values, rot=0)
sns.barplot(x=industry_delivery_totals.index, y=industry_delivery_totals.values, color='skyblue')
plt.title('나이별 고객수')
plt.xlabel('고객 나이')
plt.ylabel('수')
plt.xticks(rotation=0)
plt.show()
```

```python
bins = pd.cut(data['Avg_Open_To_Buy'], bins=range(0, 10001, 1000))

# Group by the bins and calculate the sum of 'CLIENTNUM' in each bin
industry_delivery_totals = data.groupby(bins)['CLIENTNUM'].sum().sort_values(ascending=False)

# Plotting the bar graph
plt.figure(figsize=(12, 4))
sns.barplot(x=industry_delivery_totals.index, y=industry_delivery_totals.values, color='skyblue')
plt.title('평균 소비 금액별 고객수')
plt.xlabel('평균 소비 금액 구간')
plt.ylabel('수')
plt.show()
```
