K대학교에서는 다음과 같은 기준으로 상대평가하여 학점을 매긴다.

어떤 전공 강의를 듣는 학생들 45명의 점수가 다음과 같이 산출되었다.

````python
quiz_3_data
````

#### (1)
각 등급에서 가장 낮은 점수를 가진 학생을 '수문장'이라고 부른다고 하자. A등급과 B등급의 '수문장'의 점수를 구하시오.

````python
df = pd.DataFrame({'점수' : sorted(quiz_3_data)})
df.head()
````

````python
# 상위 35%에 해당하는 점수 기준
A = np.percentile(df['점수'], 65)

# 중간 35%에 해당하는 점수 기준
B = np.percentile(df['점수'], 30)
````

````python
# 등급 부여 함수
def assign_grade(score):
    if score >= A:
        return 'A'
    elif score >= B:
        return 'B'
    else:
        return 'C'

# '등급' 컬럼 추가
df['등급'] = df['점수'].apply(assign_grade)
````

````python
df
````

````python
df[df['등급']=='A']
````

````python
quiz_3_1_answers = {
    'A등급 수문장': df[df['등급']=='A'].min()['점수'],
    'B등급 수문장': df[df['등급']=='B'].min()['점수']
}

for key, value in quiz_3_1_answers.items():
    print(f"{key}: {value}점")
````

#### (2)
응급한 전염병의 사유로 정상적인 수업 진행이 어려워지면서, K대학교에서는 상대평가를 대신해 절대평가 성적 산출을 시행하기로 하였다.

|등급|점수|
|:---:|:---:|
|A|90점 이상|
|B|80점 이상 90점 미만|
|C|70점 이상 80점 미만|
|D|60점 이상 70점 미만|
|F|60점 미만|

절대평가를 시행할 때 각 등급에 대한 도수분포표와 히스토그램을 그리시오.

````python
# 등급 부여 함수
def assign_grade(score):
    if score >= 90:
        return 'A'
    elif score >= 80:
        return 'B'
    elif score >= 70:
        return 'C'
    elif score >= 60:
        return 'D'
    else:
        return 'F'

# '등급' 컬럼 추가
df['절대 등급'] = df['점수'].apply(assign_grade)
````
